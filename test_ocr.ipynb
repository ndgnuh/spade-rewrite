```python
from detect.ocr import *
import spade.transforms as transforms
from spade.models import SpadeData
```

```python
path="../../Invoice_template/Gia_tri_gia_tang_copy/25.jpg"
model = detection_predictor(arch='db_resnet50', pretrained=True,assume_straight_pages=True)
# single_img_doc=DocumentFile.from_images(path)
# print(len(single_img_doc))
x=[]
img=cv2.imread(path)
# # print(img)

# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
x.append(img)
result=model(x)


img_copy=img.copy()
h,w,c=img.shape
bboxes=[]
for box in result[0]:
    x1=int(box[0]*w)
    y1=int(box[1]*h)
    x2=int(box[2]*w)
    y2=int(box[3]*h)
    bboxes.insert(0,[x1,x2,y1,y2])
    img_copy=bounding_box(x1,y1,x2,y2,img_copy)
plot_img(img_copy,(10,10))
raw_text=Vietocr_img(img,bboxes)
g=arrange_bbox(bboxes)
rows = arrange_row(g= g)
# st.write(raw_text)
# st.write(rows)
for row in rows:
    x=""
    for i in row:
        x=x+" "+raw_text[i]
    print(x)
SpadeData(texts=raw_text,
                     boxes=bboxes,
                     width=w,
                     height=h,
                     relations=None)
```
